```python
import os
from PIL import Image
```

```python
img_folder = os.path.join("../../","Dataset/","ZAPATOS")
preprocessed_img_folder = os.path.join("../../","Dataset/","preprocessed")
train_folder = os.path.join(preprocessed_img_folder, "train")
test_folder = os.path.join(preprocessed_img_folder, "test")
img_folder, preprocessed_img_folder, train_folder, test_folder
```

```python
imgs = os.listdir(img_folder)
imgs_location = []

for i in imgs:
    tmp = img_folder + "/" + i
    imgs_location.append(tmp)

imgs_location[:5]
```

```python
def base_model(path):
    name = os.path.splitext(os.path.basename(path))[0]
    parts = name.split('_')
    return "_".join(parts[:2])

unique = {}
for path in imgs_location:
    key = base_model(path)
    if key not in unique:
        unique[key] = path

imgs_location = list(unique.values())
```

```python
import random

random.shuffle(imgs_location)

train_imgs = imgs_location[:177]
#test_imgs = imgs_location[142:177]

for i in train_imgs:
    img = Image.open(i)
    img = img.resize((64,64))
    base = os.path.splitext(os.path.basename(i))[0]
    out_path = os.path.join(train_folder, f"{base}.png")
    img.save(out_path, format="PNG")

#for i in test_imgs:
#    img = Image.open(i)
#    img = img.resize((64,64))
#    base = os.path.splitext(os.path.basename(i))[0]
#    out_path = os.path.join(test_folder, f"{base}.png")
#    img.save(out_path, format="PNG")
```
